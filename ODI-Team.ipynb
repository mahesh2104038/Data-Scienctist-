```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url7='https://www.icc-cricket.com/rankings/mens/team-rankings/odi'
```

```python
page7=requests.get(url7)
print(page7)
```

```python
soup7=BeautifulSoup(page7.content,'lxml')
soup7
```

```python
master_container=soup7.find_all('section',{'class':'widget rankings-widget '})
print(type(master_container))
print(master_container)
```

```python
table2=soup7.find('table',{'class':'table'})
table2
```

```python
headers7=[]
for i in table2.find_all('th'):
    col_name=i.text
    headers7.append(col_name)
```

```python
headers7
```

```python
table2.find_all('tr')[1:]
```

```python
df_table2=pd.DataFrame(columns=headers7)
```

```python
for i in table2.find_all('tr')[1:]:
    row_data=i.find_all('td')
    rows=[t.text for t in row_data]
    length=len(df_table2)
    df_table2.loc[length]=rows
```

```python
df_table2
```

```python
df_table2['\nTeam\nT\n'] = df_table2['\nTeam\nT\n'].str.replace('\n', '') 
df_table2['\nRating\nR\n']=df_table2['\nRating\nR\n'].str.replace('\n','')
```

```python
url_president='https://presidentofindia.nic.in/former-presidents.htm'
```

```python
page_presedent=requests.get(url_president)
print(page_presedent)
```

```python
soup4=BeautifulSoup(page_presedent.content)
soup4
```

```python
name=[]
for i in soup4.find_all("div",{"class":"presidentListing"}):
    
    name.append(i.text.strip())
print(name)
```

```python
df = pd.DataFrame({'Name of Presedent':name})
df
```

```python
df_update=pd.DataFrame(df["Name of Presedent"].str.split('\n').fillna('[]').tolist(),
                       columns=['Name of Presedent', 'Term of Office','url_link'])
df_update
```
