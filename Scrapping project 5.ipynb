```python
#Top 10 ODI Batsmen along with the records of their team and rating.

import requests 
from bs4 import BeautifulSoup
page01=requests.get('https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting')
soup= BeautifulSoup(page01.content)
page01

#Top 10 ODL Between in men along with the records of their team and rating 

table= soup.find("table", {"class":"table rankings-table"})
table
rows=[]
#find all TR tag(rows)
for row in table:
    rows = table.find_all('tr')
#find all td tags(columns)
for tr in rows:
    data=[]
    cols = tr.find_all('td')
    for td in cols:
        data.append(td.text.strip())
    print(data)
```

```python
#Top 10 ODI teams in men’s cricket along with the records for matches, points and rating


page02=requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
soup2 = BeautifulSoup(page02.content)
page02

table = soup2.find("table",{"class":"table"})
table

rows=[]
for row in table:
    rows = table.find_all('tr')
for tr in rows:
    data=[]
    cols = tr.find_all('td')
    for td in cols:
        data.append(td.text.strip())
    print(data)
```
